### ブロッホ球による状態ベクトルの可視化

<div class="alert alert-block alert-success">
<b>このプログラムの目的：</b> ブロッホ球上で状態ベクトルを表示するコードを例示するものです。
</div>

###### 出典　https://ja.wikipedia.org/wiki/%E3%83%96%E3%83%AD%E3%83%83%E3%83%9B%E7%90%83

量子ビットの任意の状態$|\psi\rangle$は以下のように$|0\rangle$と$|1\rangle$の重ね合わせで表現できる：
$$|\psi\rangle = \cos \frac{\theta}{2}|0\rangle + e^{i\phi} \sin \frac{\theta}{2}|0\rangle$$

二つの直交する純粋状態の重ね合わせで表現できる量子状態を単位球面上に表す表記法である。  
このような球をブロッホ球という。物理学者フェリックス・ブロッホ (Felix Bloch) にちなんで名付けられた。

$|0\rangle$と$|1\rangle$はそれぞれ列ベクトルとして
$$ |0\rangle = \begin{bmatrix}1 \\ 0\end{bmatrix}, \quad |1\rangle = \begin{bmatrix}0 \\ 1\end{bmatrix}$$
であるが、内積をとるとゼロとなるため直交している。この２つのベクトルのなす角は、ブロッホ球での表示は$\pi/2$の2倍の$\pi$で表される。  

##### 出典
https://docs.quantum.ibm.com/api/qiskit/qiskit.visualization.plot_bloch_multivector    

インプットが状態ベクトルのとき、ブロッホ球で状態ベクトルを表現をするには　plot_bloch_multivector を使う。（複数量子ビットの場合も適用可）：

```python
from qiskit.quantum_info import Statevector
from qiskit.visualization import plot_bloch_multivector
import numpy as np
```

#### $ |0\rangle = \begin{bmatrix}1 \\ 0 \end{bmatrix}$ をブロッホ球上で示す

```python
# Define example statevector
psi1 = Statevector([1, 0])  # |0⟩

# Display statevector on the Bloch sphere
# 図の大きさを指定する際は、figsize オプションで指定する。
plot_bloch_multivector(psi1, figsize=(5, 5) )
```

#### $ |1\rangle = \begin{bmatrix} 0 \\ 1 \end{bmatrix}$ をブロッホ球上で示す

```python
# Define example statevector
psi2 = Statevector([0, 1])  # |1⟩

# Display statevector on the Bloch sphere
plot_bloch_multivector(psi2)
```

#### $ |+\rangle = \frac{1}{\sqrt{2}}(|0\rangle + |1\rangle) = \frac{1}{\sqrt{2}}\begin{bmatrix} 1 \\ 1 \end{bmatrix}$ をブロッホ球上で示す

```python

# Define example statevector
psi3 = Statevector([1/np.sqrt(2), 1/np.sqrt(2)])  # |+⟩

# Display statevector on the Bloch sphere
plot_bloch_multivector(psi3)
```

#### $ |-\rangle = \frac{1}{\sqrt{2}}(|0\rangle - |1\rangle) = \frac{1}{\sqrt{2}}\begin{bmatrix} 1 \\ -1 \end{bmatrix}$ をブロッホ球上で示す

```python
# Define example statevector
psi4 = Statevector([1/np.sqrt(2), -1/np.sqrt(2)])  # |-⟩

# Display statevector on the Bloch sphere
plot_bloch_multivector(psi4)
```

#### $ |i\rangle = \frac{1}{\sqrt{2}}(|0\rangle + i |1\rangle) = \frac{1}{\sqrt{2}}\begin{bmatrix} 1 \\ i \end{bmatrix}$ をブロッホ球上で示す

#### Python では、複素数$ a + b i$  ($a,b$は実数）を表わすときには、${\rm complex(a,b)}$をつかう。1j でもよい。

```python
# Define example statevector
psi5 = 1/np.sqrt(2) * (psi1 + complex(0,1) * psi2)

# Display statevector on the Bloch sphere
plot_bloch_multivector(psi5)
```

#### Python での虚数単位は$j$で表す

```python
psi5
```

#### $ |-i\rangle = \frac{1}{\sqrt{2}}(|0\rangle - i |1\rangle) = \frac{1}{\sqrt{2}}\begin{bmatrix} 1 \\ -i \end{bmatrix}$ をブロッホ球上で示す

```python
# Define example statevector
psi6 = 1/np.sqrt(2) *(psi1 - complex(0,1) * psi2)

# Display statevector on the Bloch sphere
plot_bloch_multivector(psi6)
```

```python
psi6
```

練習問題
$$ |\psi\rangle = \frac{\sqrt{3}}{2}|0\rangle + \frac{1}{2}|1\rangle $$
について、ブロッホ球で状態ベクトルを表示するコードを書き、結果を表示してください。

```python
from qiskit.quantum_info import Statevector
from qiskit.visualization import plot_bloch_multivector
import numpy as np

psi1 = Statevector([1, 0])  # |0⟩
psi2 = Statevector([0, 1])  # |1⟩

# Define example statevector
psi7 = np.sqrt(3) / 2 * psi1 + 1/2 * psi2

# Display statevector on the Bloch sphere
plot_bloch_multivector(psi7)
```

```python
psi7.is_valid()
```

練習問題
 $$ \frac{2}{3} |0\rangle + \frac{1-2i}{3}|1\rangle$$
 
について、ブロッホ球で状態ベクトルを表示するコードを書き、結果を表示してください。

```python
from qiskit.quantum_info import Statevector
from qiskit.visualization import plot_bloch_multivector
import numpy as np

psi1 = Statevector([1, 0])  # |0⟩
psi2 = Statevector([0, 1])  # |1⟩

# Define example statevector
psi8 = 2/3 * psi1 + (1-2j)/3 * psi2

# Display statevector on the Bloch sphere
plot_bloch_multivector(psi8)
```

```python
psi8.is_valid()
```

練習問題
 
$$ |\psi\rangle = \frac{1}{\sqrt{2}}(|0\rangle + e^{\frac{\pi i}{6}}|1\rangle) $$

について、状態ベクトルのブロッホ球で状態ベクトルを表示するコードを書き、結果を表示してください。

```python
from qiskit.quantum_info import Statevector
from qiskit.visualization import plot_bloch_multivector
import numpy as np

psi1 = Statevector([1, 0])  # |0⟩
psi2 = Statevector([0, 1])  # |1⟩

# Define example statevector
psi9 = 1/np.sqrt(2) * (psi1 + np.exp(np.pi * 1j /6) * psi2)

# Display statevector on the Bloch sphere
plot_bloch_multivector(psi9)
```

```python
psi9.is_valid()
```
