### ブロッホ球上で状態ベクトルの図示する方法について
　　　インプットが状態ベクトル（複数量子ビットも可）：　plot_bloch_multivector  
　　　インプットがCartesianまたは極座標ベクトル：　plot_bloch_vector


##### 出典
https://docs.quantum.ibm.com/api/qiskit/qiskit.visualization.plot_bloch_multivector    https://docs.quantum.ibm.com/api/qiskit/qiskit.visualization.plot_bloch_vector

```python
from qiskit.circuit import QuantumCircuit
from qiskit.quantum_info import Statevector
from qiskit.visualization import plot_bloch_multivector
from qiskit.visualization import plot_bloch_vector
import numpy as np
```

#### 状態ベクトルでのインプットとなる場合

```python
state = Statevector([1/np.sqrt(2), 1/np.sqrt(2)])
plot_bloch_multivector(state)
```

#### ブロッホ球をファイル出力する場合

```python
plot_bloch_multivector(state, filename='./Bloch1.png')
```

```python
plot_bloch_multivector(state, filename='./Bloch1.jpg')
```

####　Cartesian でのインプットとなる場合

```python
plot_bloch_vector([1,0,0], title="New Bloch Sphere")
```

```python
# 極座標　spherical coordinates
plot_bloch_vector([1, np.pi/2, np.pi/3], coord_type='spherical')
```
